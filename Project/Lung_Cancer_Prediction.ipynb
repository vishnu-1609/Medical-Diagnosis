#### Import required libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

#### Reading the dataset

```python
data=pd.read_csv("C:/Users/User/Documents/Medical_Diagnosis_System/project/Datasets/survey lung cancer.csv")
```

```python
data["LUNG_CANCER"]=data["LUNG_CANCER"].map({"YES":1,"NO":0})
data
```

```python
data["binaryClass"].value_counts()
```

##### 0 ----> Does not have Lung Cancer
##### 1 ----> Has Lung Cancer

```python
data["GENDER"]=data["GENDER"].map({"M":0,"F":1})
data
```

#### Splitting independent variables and dependent variable

```python
X=data.drop("LUNG_CANCER",axis=1)
Y=data["LUNG_CANCER"]
data
```

#### Splitting training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=7)
```

#### Model Training

```python
lmodel=LogisticRegression(max_iter=712)
lmodel.fit(X_train.values,Y_train.values)
```

#### Model Evaluation

```python
Train_data_predict=lmodel.predict(X_train)
Train_data_accuracy=accuracy_score(Y_train,Train_data_predict)
print("Train data accuracy:",Train_data_accuracy)
Test_data_predict=lmodel.predict(X_test)
Test_data_accuracy=accuracy_score(Y_test,Test_data_predict)
print("Test data accuracy:",Test_data_accuracy)
```

#### Predictive System

```python
input_data = (0,61,1,1,1,1,2,2,1,1,1,1,2,1,1)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = lmodel.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Lung Cancer')
else:
  print('The Person has Lung Cancer')
```
