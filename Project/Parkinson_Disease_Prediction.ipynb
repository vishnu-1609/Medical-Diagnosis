#### Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
import pickle
```

#### Dataset collection


```python
data=pd.read_csv("./Datasets/parkinson_data.csv")
data.head()
```

#### Separating independent variables and target variable

```python
# independent variables
X=data.drop(columns=["name","status"],axis=1) 
# dependent variable
Y=data["status"]
```

```python
X
```

```python
Y
```

#### Splitting training and testing data


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=7)
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

#### Data standardization

```python
# scaler = StandardScaler()
# scaler.fit(X_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
# print(X_train)
```

 #### Support Vector Machine


```python
# training a model on training data
model=svm.SVC(kernel='linear')
model.fit(X_train,Y_train)
```

#### Model evaluation

```python
x_train_predict=model.predict(X_train)
x_train_predict
x_test_predict=model.predict(X_test)
x_test_predict
```

```python
Y_test
```

##### Accuracy score

```python
train_data_accuracy=accuracy_score(Y_train,x_train_predict)
print("Train data accuracy:",train_data_accuracy)
test_data_accuracy=accuracy_score(Y_test,x_test_predict)
print("Test data accuracy:",test_data_accuracy)
```

#### Predictive system
